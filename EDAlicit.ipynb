# Base licitacoes.csv

```python
licit = pd.read_csv('Bases/licitacoes.csv')
```

```python
licit.shape
```

```python
licit.drop(columns=['link'], inplace=True)
licit['abertura_sessao'] = pd.to_datetime(licit['abertura_sessao'], errors= 'coerce').dt.date
colunas_categoria = ['objeto', 'modalidade', 'central_de_compras']
for col in colunas_categoria:
    licit[col] = licit[col].astype('category')
```

```python
licit
```

```python
duplicatas_edital = licit['edital'].duplicated().sum()
duplicatas_processo = licit['processo'].duplicated().sum()

print(f'Duplicatas na coluna edital: {duplicatas_edital}')
print(f'Duplicatas na coluna processo: {duplicatas_processo}')
```

```python
licit['data_publicacao'] = pd.to_datetime(licit['data_publicacao'], errors='coerce').dt.date
```

```python
licit['modalidade'].value_counts().plot(kind='bar', figsize=(10, 5))
plt.title('Quantidade de Licitações por Modalidade')
plt.xlabel('Modalidade')
plt.ylabel('Quantidade')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
licit['central_de_compras'].value_counts().head(10).plot(kind='bar', figsize=(10, 5))
plt.title('Top 10 Centrais de Compras por Quantidade de Licitações')
plt.xlabel('Central de Compras')
plt.ylabel('Quantidade')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
licit['abertura_sessao'] = pd.to_datetime(licit['abertura_sessao'], errors='coerce')
licit['mes_abertura'] = licit['abertura_sessao'].dt.month

meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
frequencia_mes = licit['mes_abertura'].value_counts().sort_index()
frequencia_mes.index = [meses[i - 1] for i in frequencia_mes.index]

frequencia_mes.plot(kind='bar', figsize=(10, 5))
plt.title('Quantidade de Licitações por Mês de Abertura')
plt.xlabel('Mês')
plt.ylabel('Quantidade')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

# Dicionário de Colunas — pessoa-categoria.csv

## 🧾 Identificação da pessoa e documentos

| Coluna                   | Significado estimado                                                                 |
|--------------------------|---------------------------------------------------------------------------------------|
| `CD_ORGAO`               | Código do órgão responsável pela licitação/compra                                    |
| `TP_DOCUMENTO`           | Tipo de documento do fornecedor (ex: CPF, CNPJ)                                      |
| `NR_DOCUMENTO`           | Número do documento (CPF ou CNPJ)                                                    |
| `TP_PESSOA`              | Tipo da pessoa (Física ou Jurídica)                                                  |
| `NM_PESSOA`              | Nome do fornecedor ou empresa                                                        |
| `DS_OBJETO_SOCIAL`       | Descrição da atividade principal do fornecedor                                       |
| `NR_INSCRICAO_ESTADUAL`  | Número da inscrição estadual                                                         |
| `NR_INSCRICAO_MUNICIPAL` | Número da inscrição municipal                                                        |

## 👩‍⚕️ Dados profissionais

| Coluna                        | Significado estimado                                                             |
|-------------------------------|----------------------------------------------------------------------------------|
| `CD_TIPO_CONSELHO_PROFISSIONAL` | Código do tipo de conselho (ex: CREA, CRM, OAB)                                  |
| `NR_CONSELHO_PROFISSIONAL`   | Número de registro no conselho profissional                                      |
| `SG_UF_CONSELHO_PROFISSIONAL` | Estado do conselho profissional                                                  |

## 📧 Contatos e internet

| Coluna             | Significado estimado                                           |
|--------------------|----------------------------------------------------------------|
| `DS_EMAIL`         | Endereço de e-mail do fornecedor                               |
| `DS_PAGINA_INTERNET` | Site ou página oficial                                        |

## 📍 Localização

| Coluna             | Significado estimado                                            |
|--------------------|-----------------------------------------------------------------|
| `SG_UF`            | Sigla da Unidade Federativa                                     |
| `CD_MUNICIPIO_IBGE`| Código IBGE do município                                        |
| `LOGRADOURO`       | Nome da rua/avenida                                             |
| `NR_ENDERECO`      | Número do endereço                                              |
| `COMPLEMENTO`      | Complemento do endereço                                         |
| `BAIRRO`           | Bairro                                                          |
| `CEP`              | Código Postal                                                   |
| `TELEFONE`         | Número de telefone                                              |

## 📑 Identificação da licitação

| Coluna              | Significado estimado                                           |
|---------------------|----------------------------------------------------------------|
| `NR_LICITACAO`      | Número da licitação                                            |
| `ANO_LICITACAO`     | Ano da licitação                                               |
| `CD_TIPO_MODALIDADE`| Código da modalidade de licitação                              |
| `NR_LOTE`           | Número do lote da licitação                                    |
| `NR_ITEM`           | Número do item                                                 |
| `NR_ITEM_ORIGINAL`  | Número original do item (antes de eventuais alterações)        |
| `DS_ITEM`           | Descrição detalhada do item licitado                           |
| `QT_ITENS`          | Quantidade de itens                                            |
| `SG_UNIDADE_MEDIDA` | Sigla da unidade de medida (ex: UN, M², KG)                    |

## 💰 Orçamento estimado

| Coluna                     | Significado estimado                                      |
|----------------------------|-----------------------------------------------------------|
| `VL_UNITARIO_ESTIMADO`     | Valor unitário estimado do item                           |
| `VL_TOTAL_ESTIMADO`        | Valor total estimado (unitário × quantidade)              |
| `DT_REF_VALOR_ESTIMADO`    | Data de referência do valor estimado                      |
| `PC_BDI_ESTIMADO`          | Percentual de Benefício e Despesas Indiretas estimado     |
| `PC_ENCARGOS_SOCIAIS_ESTIMADO` | Encargos sociais previstos                              |

## 💵 Valores homologados

| Coluna                     | Significado estimado                                         |
|----------------------------|--------------------------------------------------------------|
| `VL_UNITARIO_HOMOLOGADO`   | Valor unitário homologado (vencedor)                         |
| `VL_TOTAL_HOMOLOGADO`      | Valor total homologado                                       |
| `PC_BDI_HOMOLOGADO`        | Percentual de BDI homologado                                 |
| `PC_ENCARGOS_SOCIAIS_HOMOLOGADO` | Encargos sociais aplicados no valor homologado            |

## 📊 Classificações e categorias de orçamento

| Coluna                   | Significado estimado                                       |
|--------------------------|------------------------------------------------------------|
| `TP_ORCAMENTO`           | Tipo de orçamento (ex: global, estimativo, etc.)           |
| `CD_TIPO_FAMILIA`        | Código da família de itens                                 |
| `CD_TIPO_SUBFAMILIA`     | Código da subfamília de itens                              |
| `TP_BENEFICIO_MICRO_EPP` | Indicador se houve benefício para ME/EPP (microempresa)    |
| `PC_TX_ESTIMADA`         | Percentual de taxa estimada                                |
| `PC_IX_HOMOLOGADA`       | Percentual de índice (ou índice aplicado na homologação)   |

## 🧪 Indicadores específicos

| Coluna          | Significado estimado                                         |
|------------------|--------------------------------------------------------------|
| `BL_COVID19`     | Indicador binário se a licitação está relacionada à COVID-19 |
| `ID_SUBCAT`      | Código da subcategoria do item                               |

```python
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.ticker as mticker


valores = pd.to_numeric(df_participacoes['VL_TOTAL_ESTIMADO'], errors='coerce').dropna()
valores = valores[valores > 0]

# Limite no percentil 95
limite = valores.quantile(0.95)
valores_filtrados = valores[valores <= limite]

sns.set(style="whitegrid")
plt.figure(figsize=(8, 1.8))
sns.boxplot(
    x=valores_filtrados,
    color="#5DADE2",
    width=0.5,
    fliersize=1.5,
    linewidth=1.2
)

plt.gca().xaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'{int(x):,}'.replace(',', '.')))
plt.title('Boxplot do Valor Total Estimado', fontsize=12)
plt.xlabel('VL_TOTAL_ESTIMADO')
plt.yticks([])
plt.tight_layout()
plt.show()
```

```python
valores = pd.to_numeric(df_participacoes['VL_TOTAL_ESTIMADO'], errors='coerce').dropna()
valores = valores[valores > 0]

# Corta os outliers no percentil 95 
limite = valores.quantile(0.95)
valores_filtrados = valores[valores <= limite]

sns.set(style="whitegrid")
plt.figure(figsize=(8, 1.8))
sns.boxplot(
    x=valores_filtrados,
    color="#58D68D",  
    width=0.5,
    fliersize=0,
    linewidth=1.2,
    showfliers=False
)

plt.gca().xaxis.set_major_formatter(mticker.FuncFormatter(lambda x, _: f'{int(x):,}'.replace(',', '.')))
plt.title('Boxplot do Valor Total Estimado (até P95)', fontsize=12)
plt.xlabel('VL_TOTAL_ESTIMADO')
plt.yticks([])
plt.tight_layout()
plt.show()
```
